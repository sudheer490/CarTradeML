```python
import pandas as pd
df = pd.read_csv('app/data/recommed.csv')
df
```

```python
df = df[~df.cd_mileage_miles.isna()]
```

```python
df['cd_mileage_miles'] = df.cd_mileage_miles.astype(float)
df['year'] = df.cd_year.astype(int)
```

```python
df['cd_doors'] = df.cd_doors.map(lambda x: int(x.split(' ')[0]))
```

```python
import re
def get_mpg(m):
    pattern = r'\d+' 
    match = re.findall(pattern, m)
    if match:
        return match
    else:
        return [0,0]
```

```python
df['mpg_city'] = df['cd_mpg'].apply(lambda x: get_mpg(x)[0])
df['mpg_hwy'] = df['cd_mpg'].apply(lambda x: get_mpg(x)[1])
```

```python
df_ts = df[['cd_make','cd_model','cd_tank_size']].sort_values(by='cd_tank_size').drop_duplicates(['cd_make','cd_model'])
```

```python
df = pd.merge(df,df_ts,on=['cd_make','cd_model'])
```

```python
df['cd_tanksize'] = df[['cd_tank_size_x','cd_tank_size_y']].apply(lambda x: x[0] if pd.notna(x[0]) else x[1],axis=1)
```

```python
df.columns
```

```python
df = df[['cd_id','cd_mileage_miles','cd_year','cd_car_price','cd_make', 'cd_model', 'cd_body_style', 'cd_doors',
       'cd_engine', 'cd_transmission', 'cd_drive_type', 'cd_fuel',  'mpg_city', 'mpg_hwy', 'cd_tanksize']]
```

```python
df
```

```python
def get_tanksize(tank):
    try:
        return float(tank.split(' ')[0])
    except:
        return 0
```

```python
df['cd_tanksize'] = df.cd_tanksize.map(lambda x: get_tanksize(x))
```

```python
from sklearn.preprocessing import OneHotEncoder



# Create an instance of the OneHotEncoder
encoder_make = OneHotEncoder(sparse_output=False)  # Set sparse=False to get a dense matrix

# Fit and transform the data
makes_enc = pd.DataFrame(encoder_make.fit_transform(df[['cd_make']]))
makes_enc.columns = encoder_make.get_feature_names_out()

encoder_model = OneHotEncoder(sparse_output=False)  # Set sparse=False to get a dense matrix

# Fit and transform the data
model_enc = pd.DataFrame(encoder_model.fit_transform(df[['cd_model']]))
model_enc.columns = encoder_model.get_feature_names_out()

encoder_engine = OneHotEncoder(sparse_output=False)
engine_enc = pd.DataFrame(encoder_engine.fit_transform(df[['cd_engine']]))
engine_enc.columns = encoder_engine.get_feature_names_out()

encoder_transmission = OneHotEncoder(sparse_output=False)
transmission_enc = pd.DataFrame(encoder_transmission.fit_transform(df[['cd_transmission']]))
transmission_enc.columns = encoder_transmission.get_feature_names_out()

encoder_drivetype = OneHotEncoder(sparse_output=False)
drivetype_enc = pd.DataFrame(encoder_drivetype.fit_transform(df[['cd_drive_type']]))
drivetype_enc.columns = encoder_drivetype.get_feature_names_out()

encoder_body_style = OneHotEncoder(sparse_output=False)
body_style_enc = pd.DataFrame(encoder_body_style.fit_transform(df[['cd_body_style']]))
body_style_enc.columns = encoder_body_style.get_feature_names_out()

encoder_fuel = OneHotEncoder(sparse_output=False)
fuel_enc = pd.DataFrame(encoder_fuel.fit_transform(df[['cd_fuel']]))
fuel_enc.columns = encoder_fuel.get_feature_names_out()

```

```python
from sklearn.preprocessing import LabelEncoder

encoder_make = LabelEncoder()  # Set sparse=False to get a dense matrix

# Fit and transform the data
df['enc_make'] = encoder_make.fit_transform(df['cd_make'])



encoder_model = LabelEncoder()  # Set sparse=False to get a dense matrix

# Fit and transform the data
df['enc_model'] = encoder_model.fit_transform(df['cd_model'])
encoder_engine = LabelEncoder()  # Set sparse=False to get a dense matrix

# Fit and transform the data
df['enc_engine'] = encoder_engine.fit_transform(df['cd_engine'])
encoder_transmission = LabelEncoder()  # Set sparse=False to get a dense matrix

# Fit and transform the data
df['enc_transmission'] = encoder_transmission.fit_transform(df['cd_transmission'])
encoder_drive_type = LabelEncoder()  # Set sparse=False to get a dense matrix

# Fit and transform the data
df['enc_drive_type'] = encoder_drive_type.fit_transform(df['cd_drive_type'])
encoder_body_style = LabelEncoder()  # Set sparse=False to get a dense matrix

# Fit and transform the data
df['enc_body_style'] = encoder_body_style.fit_transform(df['cd_body_style'])

encoder_fuel = LabelEncoder()  # Set sparse=False to get a dense matrix

# Fit and transform the data
df['enc_make'] = encoder_fuel.fit_transform(df['cd_fuel'])
```

```python
import numpy as np
from tqdm import tqdm
from sklearn.metrics.pairwise import cosine_similarity

# Compute similarity matrix in one go
sim_matrix = cosine_similarity(al.values)

# Get top 10 similar indices for each row (excluding itself)
similiarity = np.argsort(-sim_matrix, axis=1)[:, 1:11]  # Sort in descending order

# similiarity now contains the top 10 similar indices for each row
print(similiarity)
```

```python
s = np.array(similiarity)
```

```python
sims = []
for i in tqdm(s):
    si = df.iloc[i,].cd_id.values.tolist()
    sims.append(si)
```

```python
sims = list(sims)
```

```python
b = []
for row in sims:
    a = []
    for ele in row:
        a.append(np.array(ele).item())
    b.append(a)
        
```
